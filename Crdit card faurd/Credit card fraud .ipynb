# Preparing the dataaset

    Spliting the data set into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.drop('Class',axis=1),data['Class'],test_size=0.3,random_state=0)
```

```python
#importing the stander scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Performming Logistic regresion 

```python

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter=1000)
LR.fit(x_train,y_train)
prediction=LR.predict(x_test)
confusion_mat(y_test,prediction,"Logistic Regression")
```

```python
y_pred = LR.predict(x_test)
y_pred

#making the classification report

print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
#printing the Accuracy score
print("Accuracy score:" ,accuracy_score(y_test,y_pred))
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, LR.predict_proba(x_test)[:, 1])

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

```python
# roc curve
fpr ,tpr, thresholds= roc_curve(y_test,LR.predict_proba(x_test)[:,1])
roc_auc =auc(fpr,tpr)
plt.figure()
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
